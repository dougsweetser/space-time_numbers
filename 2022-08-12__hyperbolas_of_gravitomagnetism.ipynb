```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['lines.color'] = 'k'
mpl.rcParams['axes.prop_cycle'] = mpl.cycler('color', ['k'])

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
```

```python
def axes():
    plt.axhline(0, alpha=.1)
    plt.axvline(0, alpha=.1)
    
axes()

for i in range(1, 4):
    plt.contour(x, y,(x**2/i**2 - y**2/i**2), [1], colors='b')
    plt.contour(x, y,(-x**2/i**2 + y**2/i**2), [1], colors='r')

plt.axline([0, 0], [1, 1], linestyle='--')
plt.axline([0, 0], [1, -1], linestyle='--')

axes=plt.gca()
axes.set_aspect(1)
axes.set_ylabel('time')
axes.set_xlabel('3D Space')
#plt.plot(x, y, linestyle='--', color='k', lw=1, scalex=False, scaley=False)
plt.show()
plt.savefig("hyperbolas_of_space-time.png")
```

```python
for i in range(1, 4):
    plt.contour(x, y, 2 * x * y / i**2, [1], colors='b')
    plt.contour(x, y, -(2*x * y/i**2), [1], colors='r')

plt.axline([0, 0], [1, 0],linestyle='--')
plt.axline([0, 0], [0, 1], linestyle='--')

axes=plt.gca()
axes.set_aspect(1)
axes.set_ylabel('Energy')
axes.set_xlabel('3D Momentum')
#plt.plot(x, y, linestyle='--', color='k', lw=1, scalex=False, scaley=False)
plt.show()
plt.savefig("hyperbolas_of_momentum_energy_in_g_field.png")
```
