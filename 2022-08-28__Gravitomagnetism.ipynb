## A new symmetry, a new constraint

Space-time number calculations consistently bring to the table new, well-formed terms that get dropped in the process of tensor contractions that focus on what is important (according to our educational traditions). Nature plays with all all the cards all the time. Consider the square of events in space-time:

$(t, x, y, z)^2 = (t^2 - (x^2 + y^2 + z^2), 2 t x, 2 t y, 2 t z)$

The first term in the square is the Lorentz invariant interval $\tau$. The other three as far as I know do not have a formal name. Studying this squared product would fall under conformal geometry because multiplying $(0, x, y, z)$ by the scalar $t$ cannot change angles. I call these three terms "space-times-time", a descriptive label. This is not in the standard lexicon of physics, but I like to call these three terms "Lorentz variant". We understand exactly how they transform under a Lorentz boost ($t x \rightarrow t' x' = \gamma^2 \beta_x t x$). 

Is there any physical process where the first term could be variant, but the other three terms are invariant? I don't think so because different observers would scramble the spatial 3-vector. What is one person's positive $x$ is another\'s negative $x$. If one worked in a tangent space of space-time then the process of creating a tangent space effectively subtracts away the observer. Consider a square in momentum-energy space:

$(E, P_x, P_y, P_z)^2 = (E^2 - (P_x^2 + P_y^2 + P_z^2), 2 E P_x, 2 E P_y, 2 E P_z) =  (m^2, 2 E P_x, 2 E P_y, 2 E P_z)$

The scalar term is the rest mass which is Lorentz invariant for inertial observers. The other three terms don't have a name, but I will call it momentum-times-energy. Momentum-times-energy is Lorentz variant for inertial observers just as before: $P_x E \rightarrow P_x' E' = \gamma^2 \beta_x P_x E$. Is there any branch of physics for non-inertial observers where the momentum-times-energy is invariant, but the observered mass squared is different? This is accidentally almost true for the Schwarzschild solution of general relativity. It is my proposal that this is an exact symmetry for gravity. How does one move along these hyperbolas? I want to be careful about how I talk about the speed of light _c_. As in special relativity, locally, all efforts to measure the speed of light _c_ get the same value. Imagine an experimentalist, call her Alice. Alice sets up equipment to measure the wavelength and the frequency of a stable source of light in a gravity field. The following tranformation law was first described in a YouTube comment by Purple Penguin (strange but true). Assume all the standard definitions of special relativity apply for measuring intervals of time and the distance between the ends of meter sticks. The one rule that gets dropped is that two observers have to agree that each other's measurement of the speed of light is _c_. 

Start with a coordinate transformation we know works:

$t \rightarrow t' = \beta t - \gamma \beta R$

$R \rightarrow R' = \beta R - \gamma \beta t$

where $\beta = v/c$ and $\gamma = \frac{1}{\sqrt{1 - \beta^2}}$. Our goal is to work in the tangent space of space-time, so we are after $dt$, $dt'$, $dR$ and $dR'$. As such, we can add any constant function and it will not change the deltas. The goal is to eliminate the dependence on $R$ in the first expression. Create a new coordinate transformation:

$t'' = t' + f(R')$

Let $f(R') = - \beta R'$, so

$t'' = \gamma t + \gamma \beta R - \beta R'$

$= \gamma t + \gamma \beta R - \gamma \beta R - \gamma \beta^2 t = t/\gamma$

In the tangent space for this choice of coordinates, $d t'' = \frac{1}{\gamma} dt$. What might this mean physically? If multiplied by $m /d \tau$, this would be energy. For this coordinate system choice, both observers would agree no energy is being used.

Because both $t$ and $t''$ are on the same axis, the same will be true for $R$ and $R''$. We can conclude:

$dR'' = \gamma dR$

Thus:

$\frac{d R''}{d t''} = \gamma^2 \frac{d R}{d t}$

$d R'' d t'' = d R d t$

This is the momentum-times-energy symmetry discussed in the previous section. Notice how by measuring what the speed of light measurement appears to another observer, one has an experimental handle on the escape velocity of gravity. For a complicated system, say it was electrically changed and spinning, the experimentalist might determine the answer before a theortical physicist did the calculation.

There is also a quick path to how differential elements of intervals change due to gravity.

$d \tau'' = d t''^2 - d R''^2$

$=\frac{1}{\gamma^2} dt^2 - \gamma^2 d R^2$

If one were to drop in Newton's escape velocity into this expression, while it might get the bending of light correct, it would not pass more refined weak field gravity tests. ## Deriving the Homogeneous Maxwell Equations

The no magnetic monopoles law and Faraday's law are usually presented as vector identities. There is nothing wrong with such a commonly used approach. There are often multiple perspectives on the same identity in math and physics. While this approach might be somewhere in the literature, I was fortunate enough to stumble upon it, so will share it now.

Is there another Lorentz invariant made up of only the electric and magnetic fields? I recall seeing in Jackson's "Classical Mechanics" that the dot product of an electric and magnetic field was invariant under a Lorentz boost.

Start with the sum and difference of the electric and magnetic fields. The norm squared of each should be the same because the only difference is which direction the magnetic field points in, not the size of that field. The electric fields are identical. Here is the norm squared for both:

$$||(g, \vec{B} - \vec{E})||^2 = (g^2 + (B - E)^2, \vec{0}) = (g^2 + B^2 - 2 E \cdot B + E^2, \vec{0}))$$

$$||(g, -\vec{B} - \vec{E})||^2 = (g^2 + (-B - E)^2, \vec{0}) = (g^2 + B^2 + 2 E \cdot B + E^2, \vec{0}))$$

Subtract the second from the first:

$$(- 4 E \cdot B, \vec{0}) =(0, \vec{0})$$

Notice that the scalar field $g$ is not in this expression. There will be no analog to the homogeneous Maxwell equations for gravity. This is in contrast with standard approaches to Gravito-Electromagnetism. Apply the Euler-Lagrange equations to the dot product.

```python
norm_D_potential = norm_squared(D_potential)
norm_potential_D = norm_squared(potential_D)
norm_D_potential.subs(fields).display("||D_A||²")
norm_potential_D.subs(fields).display("||A_D||²")
```

```python
norm_dif = dif(norm_D_potential, norm_potential_D)
norm_dif.subs(simpler).expand().display("diff of norms")
```

```python
norm_dif_t = euler_equations(norm_dif.t, φ(t, x, y, z), [t, x, y, z])[0]
display(norm_dif_t)
norm_dif_t.lhs.subs(simpler)
```

Here again, sympy is not "figuring it all out." By grabbing the left-hand side (lhs), the substitutions lead to a zero. Starting at the first term, one should be able to spot the vector identity: $\nabla \cdot (\nabla \times A) = 0$. This is just a different path to the no magnetic monopoles law.

Use the code that generated the Maxwell source equations to do the same for the homogeneous equations:

```python
for p in p4:
    homogeneous = euler_equations(
                    dif(norm_squared(vector_q(Dq(D_potential, [t, x, y, z], reverse=False))),
                        norm_squared(vector_q(Dq(D_potential, [t, x, y, z], reverse=True)))).t, p)[0]
    display(homogeneous)
    print(homogeneous.lhs.subs(simpler))
```

It will require considerable effort to spot Faraday's law:

$$\frac{\partial B}{\partial t} + \vec{\nabla} \times \vec{E} = \vec{0}$$

It is easier to see that all the terms do cancel. I like that the all the homogeneous equations are generated together.

$$(\nabla \cdot B, \frac{\partial B}{\partial t} + \vec{\nabla} \times \vec{E}) = (0,\vec{0})$$

It should be noted that these so-called vector identities in electromagnetism do not have a gravity analog. Perhaps that is another reason why gravity only attracts.

## Symmetries Under Conjugation

Everyone is aware of the conjugate operator. In an unpublished notebook, Gauss wrote down the usual conjugate and three others. In 1999, D. Sweetser found simple expressions for these:

$$q^* = (t, -x, -y, -z)$$

$$q^{*1} \equiv (i q i)^* = (-t, x, -y, -z)$$

$$q^{*2} \equiv (j q j)^* = (-t, -x, y, -z)$$

$$q^{*3} \equiv (k q k)^* = (-t, -x, -y, z)$$

Each of these flips signs three times. There are all discrete transformations because they are involutions:

$$(q^*)^* = (q^{*1})^{*1} = (q^{*2})^{*2} = (q^{*3})^{*3} = q$$

There is no way to be "almost" a conjugate. Based on a short exchange with Aleks Kleyn, there are also involutions that could be defined that have an even number of sign changes:

$$q^{*e} \equiv i^2 q = (-t, -x, -y, -z)$$

$$q^{*1e} \equiv (i q i) = (-t, -x, y, z)$$

$$q^{*2e} \equiv (j q j) = (-t, x, -y, z)$$

$$q^{*3e} \equiv (k q k) = (-t, x, y, -z)$$

Is there a way to use these conjugates and leave the derivation of the Maxwell source and homogeneous equations invariant? Both the derivative and potential need to be hit with the same conjugates:

## Takeaways

Every approach to deriving the Maxwell equations I have seen uses the anti-symmetric rank-2 field strength tensor $F^{\mu\nu}$ which neatly puts electric and magnetic fields in their six separate places. This notebook provides an alternative.

The three fields are constructed the simplest and most direct way possible: by taking all possible space-time derivatives of a space-time potential. The first field _g_ is born Lorentz invariant. The other two are well-known ways for representing a difference between the magnetic and electric fields (_B_ - _E_). That difference is not invariant, but with a small bit of effort it was easy enough to construct a Lorentz invariant Lagrange density composed of three squared fields. This is the heart of the proposal which need only be analyzed.

Even before doing any calculations, it is interesting to see a sign difference between the squared gravity field and squared electric field. The sign difference is entirely due to the mathematical structure of space-time numbers.

If one sets the gravity field effectively to zero, applying the Euler-Lagrange equations generates the Maxwell source equations, Gauss's and Ampere's law, together in one expression. Then look at an electrically neutral expression to generate four equations for gravity theory that has 16 terms in all. These differential equations appear almost trivial compared to what happens in general relativity. 

The simplest possible field equation is only a function of time. That implies Newton's law of gravity would also have to be a function of time. Newton wrote the law down as an inverse distance-between-masses squared expression. Relativity has taught us to think of time and space as having a deep relationship. If the distance-between-masses is viewed as the time it takes to travel between two masses, the great riddle of action at a distance is laid to rest - the law is about time of flight between masses, not distance.

Is there a symmetry principle in play? Look to general relativity for clues. The Lorentz group of Minkowski space-time was not the place to start according to general relativity. Instead one needs to work with the Poincaré group of a tangent space of space-time. Work with momentum-energy space. Accidentally, momentum-times-energy almost is invariant. It would be an interesting dual symmetry: for two inertial observers, the mass squared was invariant, but in a gravity field, two non-inertial observers would agree about momentum-times-energy. I do ask the reader to be generous in how that conformal symmetry has to be stated given the complexity of our understanding of geometry. Perhaps only for the simplest cases will it hold, but at least that is a start.

The way to generate the homogeneous Maxwell equations starts with something that must always be zero: the difference between the norm squared of the fields with the derivative on the left and the norm squared with the derivative on the right. There are no homogeneous equations for gravity whose norm remains the same. It is the dot product of the electric and magnetic wave that remains. For a free traveling EM wave, those fields are at 90 degrees to each other so the dot product is zero. It is not a well-known result, but putting the dot product of these two fields into the Euler-Lagrange equations results in the no magnetic monopoles law and Faraday's law.

The conjugate symmetries indicated conservation laws based on Emmy Noether's theorem. Because the symmetries are discrete, that means the conserved quantity is discrete.
